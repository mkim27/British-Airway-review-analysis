## Data to be cleaned:
* Non-response ratings recorded as -1

We can see that the data recorded -1 for any cases where customers chose not to give a rating, or for flights that didn't provide services like in-flight food or entertainment. These false responses need to be removed, because they drag down the average overall and category ratings.

* Overall ratings are scaled differently to category ratings

The overall ratings have a range of 1-10, whereas the subcategory ratings have a range of 1-5. This is supported by how the overall rating values are frequently higher than the averages of the subcategorical ratings.

```python
df_rev = df_rev.replace(-1, np.nan)
df_rev.head(10)
```

```python
df_rev.loc[:, 'rating'] /= 2
df_rev.head()
```

```python
df_rev.describe()
```

## Quick glance insights:

* In-flight food/beverages and Entertainment are offered in a significantly fewer number of flights, affecting the average British Airways customer less

* Cabin Staff Service and Ground Service have the highest average ratings

* Food/beverages and Entertainment have the lowest average ratings

* The 50% rating percentiles shows that Value-for-Money, Food/beverages, and Entertainment were the worst parts of the flight for many customers

```python
df_rev.to_csv('ba_reviews_cleaned1.csv', index=False)
```
